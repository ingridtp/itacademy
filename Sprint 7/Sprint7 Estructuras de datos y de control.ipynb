    Calculadora del índice de masa corporal

    Escribe una función que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deberá ingresar estos datos. Puedes obtener más información de su cálculo en:

    -> Índice de masa corporal IMC que es y cómo se calcula.

    La función debe clasificar el resultado en sus respectivas categorías


```python
#Calcular el IMC
def calcular_imc(peso, talla):
    return peso / (talla ** 2)

#Clasificar el IMC
def clasificar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif imc < 25:
        return "Peso normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

#Validar entrada de número mayor que 0 
def pedir_valor_positivo(mensaje):
    while True:
        try:
            valor = float(input(mensaje))
            if valor <= 0:
                print("Error: Por favor ingresa un número mayor que 0")
            else:
                return valor
        except ValueError:
            print("Error: Por favor ingresa un número válido.")
```

```python
# Main Calculadora de IMC
def main ():
    print("\nCalculadora de IMC")

    peso = pedir_valor_positivo("Ingresa tu peso en kg: ")
    talla = pedir_valor_positivo("Ingresa tu altura en m: ")

    imc = calcular_imc(peso, talla)
    clasificacion = clasificar_imc(imc)

    print(f"Tu IMC es: {imc:.2f}")
    print(f"Clasificación: {clasificacion}")

main()
```

- Ejercicio 2
Conversor de temperaturas.

    Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re). Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como mínimo, dos conversiones.

```python
#Convertir de °C a °F
def c_a_f(c):
    return (c * 9/5) + 32

#Convertir de °C a K
def c_a_k(c):
    return c + 273.15

#Conventir de °F a °C
def f_a_c(f):
    return (f - 32) * 5/9

#Convertir de °F a K
def f_a_k(f):
    return (f - 32) * 5/9 + 273.15

#Convertir de K a °C
def k_a_c(temp_k):
    return temp_k - 273.15

#Convertir de K a °F
def k_a_f(temp_k):
    return (temp_k - 273.15) * 9/5 + 32

#Validar entrada de número
def pedir_valor(mensaje):
    while True:
        try:
            temp = float(input(mensaje))
            return temp
        except ValueError:
            print("Error: Por favor ingresa un número válido.")

# Mostrar menú del conversor de temperaturas
def mostrar_menu_temperaturas():
    menu = """
Conversor de Temperatura
Selecciona la unidad de temperatura que vas a ingresar:
1. Celsius (°C)
2. Fahrenheit (°F)
3. Kelvin (K)
4. Salir
"""
    print(menu)
    return input("Ingresa el numero de tu elección: ")

# Convertir
def convertir_temperatura(opcion, valor):
    if opcion == "1": 
        f = c_a_f(valor)
        k = c_a_k(valor)
        return (valor, "°C"), (f, "°F"), (k, "K")
    elif opcion == "2":  
        c = f_a_c(valor)
        k = f_a_k(valor)
        return (valor, "°F"), (c, "°C"), (k, "K")
    elif  opcion == "3": 
        c = k_a_c(valor)
        f = k_a_f(valor)
        return (valor, "K"), (c, "°C"), (f, "°F")
    else:
        return None, None, None

#Mostrar las conversiones
def mostrar_conversiones(tuple_a, tuple_b, tuple_c):
    print(f"{tuple_a[0]:.2f} {tuple_a[1]} = {tuple_b[0]:.2f} {tuple_b[1]} = {tuple_c[0]:.2f} {tuple_c[1]}")
```

```python
#Main Conversor de temperatura
def main():
    while True:
        opcion = mostrar_menu_temperaturas()
        
        if opcion in ("1", "2", "3"):
            valor = pedir_valor("Ingresa la temperatura: ")
            
            tuple_origen, tuple_conv_1, tuple_conv_2 = convertir_temperatura(opcion, valor)
            mostrar_conversiones(tuple_origen, tuple_conv_1, tuple_conv_2)
            
        elif opcion == "4":
            print("Gracias")
            break
        else:
            print("Opción inválida. Intenta de nuevo.\n")

main()
```

- Ejercicio 3
Contador de palabras de un texto.

    Escribe una función que dado un texto, muestre las veces que aparece cada palabra.

"HOLA MUNDO. ¡Hola mundo! Habemus contador y habemus mundo también."

```python
import re
from collections import Counter

# Listado de palabras
def listar_palabras(texto):
    return re.findall(r'\b[a-záéíóúüñ]+\b', texto.lower())

#Contar total de palabras
def contar_palabras(list_a):
    return len(list_a)

#Contar palabras unicas
def contar_unicas(list_a):
    return len(set(list_a))

# Contar frecuencia de cada palabra
def mapear_frecuencias(list_a):
    return dict(Counter(list_a))

# Imprimir diccionario de manera legible para un usuario comun
def imprimir_diccionario(dict_a):
    for clave, valor in dict_a.items():
        print(f"{clave}: {valor}")
```

```python
#Main Contador de palabras
def main ():
    print("Contador de palabras")
    texto = input("Introduce un texto: ")

    list_palabras = listar_palabras(texto)
    total = contar_palabras(list_palabras)
    unicas = contar_unicas(list_palabras)
    dict_frecuencias = mapear_frecuencias(list_palabras)

    print(f"\nCantidad total de palabras: {total}")
    print(f"Cantidad de palabras únicas: {unicas}\n")
    print("Frecuencia de cada palabra:")
    imprimir_diccionario(dict_frecuencias)

main()
```

- Ejercicio 4
Diccionario inverso.

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiados las claves y valores. Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función debería imprimir un mensaje de advertencia.

```python
# Encuestas (diccionarios de ejemplos)
dict_encuesta_2021 = {
    "Laura": "Pizza",
    "Carlos": "Pasta",
    "Marta": "Ensalada"
}

dict_encuesta_2022 = {
    "Laura": "Pizza",
    "Carlos": "Pizza",
    "Marta": "Ensalada",
    "Sergio": "Pizza",
    "Isabella": "Carne",
    "Ana": "Ensalada"
}
```

```python
#Verificar valores unicos de un diccionario
def verificar_unicos(dict_a):
    list_valores = list(dict_a.values())
    return len(list_valores) == len(set(list_valores))

#Invertir diccionario con valores unicos
def invertir_diccionario(dict_a):
    return {valor: clave for clave, valor in dict_a.items()}

#Seleccionador de encuesta
def seleccionar_encuesta():
    while True:
        print("Selecciona la encuesta a invertir:")
        print("1. Encuesta 2021")
        print("2. Encuesta 2022")

        opcion = input("Ingresa 1 o 2: ")

        if opcion == "1":
            return dict_encuesta_2021
        elif opcion == "2":
            return dict_encuesta_2022
        else:
            print("Opción no válida. Intenta nuevamente.\n")          
```

```python
#Main Inversor de diccionarios con valores únicos
def main():
    dict_encuesta = seleccionar_encuesta()

    if verificar_unicos(dict_encuesta):
        dict_invertido = invertir_diccionario(dict_encuesta)
        print("Diccionario invertido:")
        imprimir_diccionario(dict_invertido)
        
    else:
        print("Error: Valores repetidos. No se puede invertir el diccionario.")

main()
```

Ejercicio 1
Diccionario inverso con duplicados

Continuando con el ejercicio 4 del nivel 1: al cliente se olvidó de comentar un detalle y resulta que los valores en el diccionario original pueden duplicarse y más, por lo que las claves intercambiadas pueden tener duplicados. En este caso, en el ejercicio anterior imprimías un mensaje de advertencia; de ser una lista.

```python
# Invertir diccionario permitiendo valores duplicadas.
def invertir_diccionario_pro(dict_a):
    dict_invertido = {}
    for clave, valor in dict_a.items():
        if valor in dict_invertido:
            dict_invertido[valor].append(clave)
        else:
            dict_invertido[valor] = [clave]
    return dict_invertido

# Simplificar diccionario que tienen valores con listas de un solo elemento.
def simplificar_valores(dict_a):
    dict_simple = {}
    for clave, valor in dict_a.items():
        if isinstance(valor, list) and len(valor) == 1:
            dict_simple[clave] = valor[0]
        else:
            dict_simple[clave] = valor
    return dict_simple
```

```python
#Main Inversor de diccionarios

def main ():
    dict_encuesta = seleccionar_encuesta()
    dict_invertido = invertir_diccionario_pro(dict_encuesta)
    dict_final = simplificar_valores(dict_invertido)

    print("Diccionario invertido:")
    imprimir_diccionario(dict_final)

main()
```

Ejercicio 2
Conversión de tipos de datos

El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estarán todos los elementos que pudieron convertirse en flotantes y la otra donde están los elementos que no pudieron convertirse. Ejemplo de la lista que recibe el cliente: list_datos = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]

```python
#Lista de datos del cliente
list_datos = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]
```

```python
def separar_flotantes(list_a):
    list_flotantes = []
    list_no_convertibles = []

    for elemento in list_a:
        if isinstance(elemento, (list, tuple)):
            list_flotantes_a, list_no_convertibles_a = separar_flotantes(elemento)
            list_flotantes.extend(list_flotantes_a)
            list_no_convertibles.extend(list_no_convertibles_a)
        else:
            try:
                list_flotantes.append(float(elemento))
            except (ValueError, TypeError):
                list_no_convertibles.append(elemento)
    return list_flotantes, list_no_convertibles
```

```python
def main():
    list_flotantes, list_no_convertibles = separar_flotantes(list_datos)

    # Mostrar los resultados
    print("Elementos convertidos a float:", list_flotantes)
    print("Elementos no convertibles a float:", list_no_convertibles)

main()
```

Contador y ordenador de palabras de un texto.

El cliente quedó contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario según la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a Z. 

```python
#Ruta del archivo
ruta_archivo = r"D:\Trabajo\Especializacion DATA - BA\Recursos\tu_me_quieres_blanca.txt"
```

```python
#Leer archivo
def leer_archivo(ruta):
    with open(ruta, 'r', encoding='utf-8') as archivo:
        return archivo.read()

#Anidar diccionario en 2 niveles, agrupando palabras por su letra inicial
def anidar_diccionario(dict_frecuencias):
    dict_anidado = {}
    for palabra, frecuencia in dict_frecuencias.items():
        letra_inicial = palabra[0]
        if letra_inicial not in dict_anidado:
            dict_anidado[letra_inicial] = {}
        dict_anidado[letra_inicial][palabra] = frecuencia
    return dict_anidado
    
#Ordenar un diccionario plano por sus claves.
def ordenar_diccionario(dict_a):
    return dict(sorted(dict_a.items()))

#Ordenar un diccionario anidado de 2 niveles
def ordenar_diccionario_anidado(dict_anidado):
    dict_ordenado = {}
    for clave, dict_sub in dict_anidado.items():
        dict_ordenado[clave] = ordenar_diccionario(dict_sub)  
    return ordenar_diccionario(dict_ordenado)

#Imprimir diccionarios anidados en formato legible para el usuario común
def imprimir_diccionario_anidado(dict_anidado):
    for letra, dict_palabras in dict_anidado.items():
        print(f"'{letra}':")
        for palabra, frecuencia in dict_palabras.items():
            print(f"  '{palabra}': {frecuencia}")
```

```python
def main():
    texto = leer_archivo(ruta_archivo)
    list_palabras = listar_palabras(texto)
    dict_frecuencias = mapear_frecuencias(list_palabras)
    dict_anidado = anidar_diccionario(dict_frecuencias)
    dict_ordenado = ordenar_diccionario_anidado(dict_anidado)

    imprimir_diccionario_anidado(dict_ordenado)

main()
```
