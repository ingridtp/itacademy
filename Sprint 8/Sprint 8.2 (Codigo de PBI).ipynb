```python
# The following code to create a dataframe and remove duplicated rows is always executed and acts as a preamble for your script: 

# dataset = pandas.DataFrame(tran_id, tran_decline, comp_id, comp_country)
# dataset = dataset.drop_duplicates()

# Paste or type your script code here:

# Nivel 1. Ejercicio 5: Dos variables categóricas.


# Importaciones 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MultipleLocator # Para definir intervalos fijos en los ejes


# Colores base
color_columna = "#2e5068"
color_borde = "#0f2a3b"
color_fondo = "#f7f9fc"

#Colores específicos
color_columna_contraste = "#7a9bb0"

# Dataset
dataset = dataset.groupby(["comp_country", "tran_decline"]).size().unstack(fill_value=0)
dataset = dataset.div(dataset.sum(axis=1), axis=0) * 100
dataset = dataset.rename(columns={False: "Aprobada", True: "Rechazada"})
dataset = dataset.sort_values("Aprobada", ascending=True)

# Figura y ejes
fig, ax = plt.subplots(figsize=(5, 5))

# Gráfico de barras apiladas
dataset.plot(
    kind='barh',
    stacked=True,
    color=[color_columna, color_columna_contraste],
    edgecolor=color_borde,
    ax=ax
)

# Estilo base
ax.set_facecolor(color_fondo)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color(color_borde)
ax.spines['left'].set_color(color_borde)
ax.tick_params(axis='x', colors=color_borde)
ax.tick_params(axis='y', colors=color_borde)

# Ajustes específicos
ax.set_xlabel("Porcentaje (%)", color=color_borde)
ax.set_ylabel("")
ax.xaxis.grid(True, linestyle='--', linewidth=0.5, color=color_borde, alpha=0.5)
ax.xaxis.set_major_locator(MultipleLocator(20))

# Leyenda
ax.legend(title="Estado", loc='upper left', bbox_to_anchor=(1,1))

plt.tight_layout()
plt.show()
```

```python
# The following code to create a dataframe and remove duplicated rows is always executed and acts as a preamble for your script: 

# dataset = pandas.DataFrame(tran_id, tran_decline, tran_amount, comp_id, comp_country)
# dataset = dataset.drop_duplicates()

# Paste or type your script code here:

# Nivel 1. Ejercicio 6: Tres variables.

# Importaciones 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MultipleLocator # Para definir intervalos fijos en los ejes

# Colores base
color_columna = "#2e5068"
color_borde = "#0f2a3b"
color_fondo = "#f7f9fc"

# Dataset 
dataset = dataset[dataset["tran_decline"] == False]
dataset = dataset.groupby('comp_country').agg({
    'tran_id': 'count',
    'tran_amount': 'mean'
}).reset_index()

dataset.columns = ['pais', 'cantidad_transacciones', 'monto_promedio']
dataset = dataset.sort_values("cantidad_transacciones", ascending=False)

# Figura y ejes
fig, ax = plt.subplots(figsize=(12, 5))

#Colores específicos
colores_por_pais = dict(zip(dataset['pais'],sns.color_palette("tab20", 15)))

# Gráfico de dispersión
sns.scatterplot(
    data=dataset,
    x='cantidad_transacciones',
    y='monto_promedio',
    hue='pais',
    palette=colores_por_pais,
    s=120,
    edgecolor=color_borde,
    ax=ax
)

# Estilo base
ax.set_facecolor(color_fondo)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color(color_borde)
ax.spines['left'].set_color(color_borde)
ax.tick_params(axis='x', colors=color_borde, labelsize=14)  
ax.tick_params(axis='y', colors=color_borde, labelsize=14)              

# Ajustes específicos
ax.set_xlabel("Cantidad de transacciones aprobadas", color=color_borde, fontsize=18)
ax.set_ylabel("Monto promedio por transacción ($)", color=color_borde, fontsize=18)
ax.xaxis.grid(True, linestyle='--', linewidth=0.5, color=color_borde, alpha=0.5)
ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color=color_borde, alpha=0.5)
ax.xaxis.set_major_locator(MultipleLocator(10))

# Leyenda con tamaño de título y texto
ax.legend(title="País", title_fontsize=10, fontsize=11, loc='upper left', bbox_to_anchor=(1.02,1))

plt.tight_layout()
plt.show()
```

Nivel 1. Ejercicio 7
Graficar un Pairplot.


```python
# The following code to create a dataframe and remove duplicated rows is always executed and acts as a preamble for your script: 

# dataset = pandas.DataFrame(tran_id, tran_decline, tran_amount, user_id, user_country)
# dataset = dataset.drop_duplicates()

# Nivel 1. Ejercicio 7: Graficar un Pairplot.

# Paste or type your script code here:
import matplotlib.pyplot as plt
import seaborn as sns

# Colores base
color_columna = "#2e5068"
color_borde = "#0f2a3b"

#Colores específicos
paleta_paises = {
    'United States': '#d62728',
    'United Kingdom': '#1f77b4',
    'Canada': '#2ca02c'    
}

# Dataset
dataset = dataset[dataset["tran_decline"] == False]

dataset = dataset.groupby(['user_id', 'user_country']).agg(
    ticket_promedio=('tran_amount', 'mean'),
    cantidad_transacciones=('tran_id', 'count')
).reset_index()

dataset = dataset.rename(columns={
    "user_country": "País de usuario",
    "ticket_promedio": "Ticket promedio",
    "cantidad_transacciones": "Cantidad de transacciones"
})

# Gráfico de pares
sns.pairplot(
    dataset, 
    vars=["Ticket promedio", "Cantidad de transacciones"],
    hue="País de usuario",
    palette=paleta_paises,
    diag_kind='kde'
)

plt.suptitle('Perfil de usuarios por país', y=1.05, color=color_borde)
plt.show()
```

```python
# The following code to create a dataframe and remove duplicated rows is always executed and acts as a preamble for your script: 

# dataset = pandas.DataFrame(tran_id, prod_id, tran_lat, tran_longitude, tran_amount, prod_price, prod_weight)
# dataset = dataset.drop_duplicates()

# Paste or type your script code here:

# Nivel 2. Ejercicio 1: Correlación de todas las variables numéricas.

# Importaciones 
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
dataset = dataset[dataset["tran_decline"] == False]

columnas_corr = ['tran_amount', 'prod_price', 'prod_weight', 'tran_lat', 'tran_longitude']
matriz = dataset[columnas_corr].corr(method='spearman')

etiquetas = ["Monto", "Precio", "Peso", "Latitud", "Longitud"]
matriz.index = etiquetas
matriz.columns = etiquetas

# Paleta de colores
color_matriz = sns.light_palette("#165f73", as_cmap=True)

# Figura
fig, ax = plt.subplots()

# Mapa de calor
sns.heatmap(matriz, annot=True, cmap=color_matriz, fmt=".2f", ax=ax)

plt.tight_layout()
plt.show()
```

Implementa un violinplot combinado con otro tipo de gráfico.

```python
# The following code to create a dataframe and remove duplicated rows is always executed and acts as a preamble for your script: 

# dataset = pandas.DataFrame(tran_id, tran_amount, tran_decline, user_id, user_country)
# dataset = dataset.drop_duplicates()

# Paste or type your script code here:

# Nivel 3. Ejercicio 1: Implementa un violinplot combinado con otro tipo de gráfico.

# Importaciones
import matplotlib.pyplot as plt
import seaborn as sns

# Colores base
color_columna = "#2e5068"
color_borde = "#0f2a3b"
color_fondo = "#f7f9fc"

# Colores específicos 
color_box = 'orange'
color_columna_contraste = "#7a9bb0"
paleta_violin = {
    "Aprobada": color_columna,
    "Rechazada": color_columna_contraste
}

#Dataset
dataset["tran_status"] = dataset["tran_decline"].map({0: "Aprobada", 1: "Rechazada"})

# Figura y eje
fig, ax = plt.subplots()

# Violin
sns.violinplot(
    data=dataset,
    x="user_country",
    y="tran_amount",
    hue="tran_status",
    split=True,
    palette=paleta_violin,
    inner=None,
    ax=ax
)

# Cajas
sns.boxplot(
    data=dataset,
    x="user_country",
    y="tran_amount",
    hue="tran_status",
    dodge=True,
    width=0.2,
    showcaps=True,
    boxprops={'facecolor':'None', 'edgecolor':color_box, 'linewidth':2},
    showfliers=False,
    whiskerprops={'color':color_box, 'linewidth':2},
    capprops={'color':color_box, 'linewidth':2},
    medianprops={'color':color_box, 'linewidth':2},
    ax=ax
)

# Sólo una leyenda
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2], title="Estado", loc="best", bbox_to_anchor=(1, 0.5))

# Estilo base
ax.set_facecolor(color_fondo)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='x', colors=color_borde, labelsize=12)
ax.tick_params(axis='y', colors=color_borde, labelsize=12)

# Ajustes específicos
ax.set_xlabel("")
ax.set_ylabel("Monto de transacción (USD)", color=color_borde, fontsize=12)
ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color=color_borde, alpha=0.5)

plt.tight_layout()
plt.show()
```

Nivel 2. Ejercicio 2
Genera un FacetGrid para visualizar múltiples aspectos de datos simultáneamente.

```python
# The following code to create a dataframe and remove duplicated rows is always executed and acts as a preamble for your script: 

# dataset = pandas.DataFrame(tran_id, tran_amount, Year, user_id, user_country)
# dataset = dataset.drop_duplicates()

# Paste or type your script code here:

# Nivel 3. Ejercicio 2: Genera un FacetGrid para visualizar múltiples aspectos de datos simultáneamente.

# Importaciones
import matplotlib.pyplot as plt
import seaborn as sns

# Colores base
color_columna = "#2e5068"
color_borde = "#0f2a3b"
color_fondo = "#f7f9fc"

# Datset
dataset = dataset[dataset["tran_decline"] == False]

# Crear FacetGrid: país en columnas, año en filas
g = sns.FacetGrid(
    dataset,
    col="user_country",
    row="Year",
    margin_titles=True,
    height=3,
    aspect=1.2
)

# Histograma por subconjunto
g.map(
    sns.histplot,
    "tran_amount",
    bins=10,
    color=color_columna,
    edgecolor=color_borde,
    alpha=0.8
)

# Estilo 
for ax in g.axes.flat:
    ax.set_facecolor(color_fondo)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_color(color_borde)
    ax.spines['left'].set_color(color_borde)
    ax.tick_params(axis='x', colors=color_borde, labelsize=11)
    ax.tick_params(axis='y', colors=color_borde, labelsize=11)
    ax.set_xlabel("Monto de transacción (USD)", color=color_borde, fontsize=11)
    ax.set_ylabel("Frecuencia", color=color_borde, fontsize=11)
    ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color=color_borde, alpha=0.5)
 
# Ajustes de títulos de filas/columnas
g.set_titles(col_template='{col_name}', row_template='{row_name}')

plt.tight_layout()
plt.show()
```
